Import Library yang diperlukan

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import datasets
import matplotlib.pyplot as plt
import pandas as pd
```

***Import Dataset***

```python
dataset = pd.read_csv('Dataset Iris.csv')
print (dataset)
```

Sebagai ilustrasi , dataset yang digunakan di dalam
program ini adalah dataset bunga IRIS yang bersumber
pada data yang dipublikasikan oleh Fisher (Fisher, 1950)
yang tersedia di website UCI Machine Learning Repository.

***Load Dataset Iris dari library scikit learn***

```python
#Load Dataset Iris dari library scikit learn
iris = datasets.load_iris()
features = iris['data']
target = iris['target']
print(target)
```

***Membuat Objek Model Decision Tree***

```python
# Membuat Objek Model Decision Tree
decisiontree = DecisionTreeClassifier(random_state=0, max_depth=None, 
                                      min_samples_split=2, min_samples_leaf=1,
                                      min_weight_fraction_leaf=0,
                                      max_leaf_nodes=None, 
                                      min_impurity_decrease=0)
```

***Training Model Decison Tree***

```python
model = decisiontree.fit(features, target)
```

***Membuat prediksi*** <br>
• Pada tahap ini dilakukan pengambilan sampel observasi dan
membuat prediksi.<br>
• Sampel yang diberikan adalah data dimensi kelopak. <br>
• Fungsi predict() digunakan untuk memeriksa kelas yang dimilikinya,
sedangkan predict_proba digunakan untuk memeriksa probabilitas
kelas dari prediksi tersebut.

```python
observation = [[5,4,3,2]]
model.predict(observation)
model.predict_proba(observation)
```
