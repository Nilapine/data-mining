# Studi Kasus : Regresi Linear Data Tinggi Badan & Berat Badan

```python
from scipy import stats
```

```python
# x data tinggi badan
x = [151, 174, 138, 186, 128, 136, 179, 163, 152, 131]
```

```python
# y data berat badan (kg)
y = [63, 81, 56, 91, 47, 57, 76, 72, 62, 48]
```

```python
slope, intercept, r, p, std_err = stats.linregress(x,y)
```

```python
#buat fungsi linear regresi => y = 0.67461045 x - 38.45508707607701
def myfunc(x):
    return slope * x + intercept
```

```python
berat_badan = myfunc(151)
```

```python
print(berat_badan)
```

# Studi Kasus : Regresi Linear Data Penjualan.xlsx

```python
#siapkan library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
#panggil dataset
dataset = pd.read_excel('Konsumen.xlsx')
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 1].values
```

```python
dataset.keys()
```

```python
dataset.shape
```

```python
#enampilkan isi sebagian dataset
dataku = pd.DataFrame(dataset)
dataku.head()
```

```python
# Split dataset menjadi training set dan testing set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=0)
```

```python
#melakukan fitting simple linear regresion pada data training
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
#memprediksi hasil test
y_pred = regressor.predict(x_test)
```

```python
#visualisasi data
plt.scatter(dataku.Gaji, dataku.Pengeluaran)
plt.xlabel("Gaji")
plt.ylabel("Pengeluaran")
plt.title("Grafik Gaji VS Pengeluaran")
plt.show()
```

```python
#visualisasi hasil prediksi pada training set
#ukuran plot
plt.figure(figsize=(10,8))

#biru adalah data observasi
plt.scatter(x_train, y_train, color='blue')

#garis merah adalah hasil prediksi dari machine learning 
plt.plot(x_train, regressor.predict(x_train), color='red')

#memberi judul dan label
plt.xlabel("Gaji")
plt.ylabel("Pengeluaran")
plt.title("Gaji terhadap Pengeluaran (training set)")

#menampilkan plot
plt.show()
```

```python
#visualisasi hasil prediksi pada test-set
#ukuran plot
plt.figure(figsize=(10,8))

#biru adalah data observasi
plt.scatter(x_test, y_test, color='blue')

#garis merah adalah hasil prediksi dari machine learning 
plt.plot(x_test, regressor.predict(x_test), color='red')

#memberi judul dan label
plt.xlabel("Gaji")
plt.ylabel("Pengeluaran")
plt.title("Gaji terhadap Pengeluaran (test set)")

#menampilkan plot
plt.show()
```
